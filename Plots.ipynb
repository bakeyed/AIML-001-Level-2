```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.5))

# Line Plot
ax1.plot(x, y1, label='sin(x)')
ax1.plot(x, y2, label='cos(x)')
ax1.set_xlabel('X-axis') #Axes lables
ax1.set_ylabel('Y-axis')
ax1.set_xlim(0, 10) #Axis limits
ax1.set_ylim(-1.5, 1.5)
ax1.legend()
ax1.set_title('Line Plot: sin(x) and cos(x)')

# Area Plot
ax2.fill_between(x, y1, alpha=0.5, label='sin(x)')
ax2.fill_between(x, y2, alpha=0.5, label='cos(x)')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_xlim(0, 10)
ax2.set_ylim(-1.5, 1.5)
ax2.legend()
ax2.set_title('Area Plot: sin(x) and cos(x)')
plt.tight_layout()
```

```python
# Creating a figure with two subplots 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.5))

# Data for the first simple bar plot
data = {'Pink': 20, 'Purple': 40, 'Black': 25, 'Brown': 60}
colors = data.keys()
values = data.values()

# Simple bar plot on ax1
ax1.bar(colors, values, color='lightblue', width=0.5)
ax1.set_title("Simple Bar Plot")

# Data for the stacked bar plot
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]

# Stacked bar plot on ax2
ax2.bar(x, y1, color='pink', label='y1')
ax2.bar(x, y2, bottom=y1, color='purple', label='y2')
ax2.set_title("Stacked Bar Plot")
ax2.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

```python
#Pie plot
labels = [
    'Be gamer',  
    'Nvidia GPU', 
    'Hate privacy',
    'Normal person'
]
sizes = [20, 15, 15, 50]

# Plotting
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.legend()
plt.title('Reasons to use Windows over Linux')
plt.show()
```

```python
# Data
data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C'], 'Value': [7, 6, 4, 5, 8, 7]}
df = pd.DataFrame(data)

# Plotting
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Box Plot')
plt.show()
```

```python
# Violin plot
sns.violinplot(x='Category', y='Value', data=df)
plt.title('Violin Plot for same data')
plt.show()
```

```python
# Margial plot
data = {'Feature1': np.random.rand(20), 'Feature2': np.random.rand(20)}
df = pd.DataFrame(data)

plt.figure(figsize=(5, 5))
sns.jointplot(x='Feature1', y='Feature2', data=df, kind='scatter', 
              marginal_kws=dict(bins=10, fill=True), 
              color="teal", height=6, ratio=4, space=0.3)

plt.suptitle('Marginal Plot', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Countrou plot

# Generate mesh grid
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plotting
plt.contour(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.title('Contour Plot Example')
plt.show()
```

```python
#Heatmap

# Sample data
data = np.array([[5, 2, 8], [9, 1, 3], [6, 7, 4]])
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap Example')
plt.show()
```
